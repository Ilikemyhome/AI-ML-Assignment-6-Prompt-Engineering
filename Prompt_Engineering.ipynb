```python
from transformers import pipeline
```

# Load the Model

```python
model = pipeline("text-generation", model="distilgpt2")
```

# Original input

```python
story = (
    "In my junior year of high school, this guy asked me on a date. "
    "He rented a Redbox movie and made a pizza. We were watching the movie and the oven beeped so the pizza was done. "
    "He looked me dead in the eye and said, “This is the worst part.” "
    "I then watched this boy open the oven and pull the pizza out with his bare hands, rack and all, screaming at the top of his lungs. "
    "We never had a second date."    
)
```

# Prompt Variations

```python
prompts = {
    "genz_role": (
        "You are a sarcastic Gen Z TikTok creator recounting a chaotic first date story to your followers. "
        "Use slang, emojis, dramatic pauses, and punchy one-liners. Make it feel like a viral TikTok voiceover.\n\n"
        f"Original story:\n{story}"
    ),
    "yoda_role": (
        "You are Yoda, the wise Jedi Master, retelling a strange tale of a youngling’s first date. "
        "Use your signature inverted sentence structure and Jedi-style reflection.\n\n"
        f"Original story:\n{story}"
    ),
    "genz_format": (
        "Rewrite the following story in the style of a Gen Z TikTok creator. "
        "Format it as a numbered list of short dramatic beats, each one like a TikTok caption.\n\n"
        f"Original story:\n{story}"
    ),
    "yoda_format": (
        "Rewrite the following story in the voice of Yoda. "
        "Format it as a series of short Jedi-style proverbs, each one numbered. "
        "Each line should reflect a moment in the story using Yoda’s syntax.\n\n"
        f"Original story:\n{story}"
    ),
    "genz_cot": (
        "First, identify the emotional beats of the story: awkward setup, unexpected behavior, shock, and decision to end it. "
        "Then rewrite the story in the voice of a sarcastic Gen Z TikTok creator, using slang, emojis, and dramatic pauses.\n\n"
        f"Original story:\n{story}"
    ),
    "yoda_cot": (
        "First, reflect on the emotional arc of the story: curiosity, confusion, pain, and wisdom gained. "
        "Then retell the story in the voice of Yoda, using inverted syntax and Jedi-style reflection.\n\n"
        f"Original story:\n{story}"
    )
}
```

# Run the prompt

```python
def run_prompt(name, prompt):
    print(f"\n--- {name.upper()} ---\n")
    response = model(prompt, max_new_tokens=300, do_sample=True, temperature=0.9)
    print(response[0]['generated_text'])
```

# Run the Code

```python
if __name__ == "__main__":
    for name, prompt in prompts.items():
        run_prompt(name, prompt)
```
